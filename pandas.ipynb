#### 2. Estrutura de dados: Series e DataFrames
* Series: É uma matriz unidimensional com rótulos(índices) - pode ser vista como uma única coluna de uma tabela
* DataFrames: É uma estrutura bidimensional (linhas e colunas), como uma planilha completa - é essencialmente um conjunto de Series

```python
# Criando uma Series
notas = pd.Series([8.5, 9.0, 7.5, 10], index=['Aluno A', 'Aluno B', 'Aluno C', 'Aluno D'])
print('Exemplo de Series: \n', notas, '\n')
```

```python
# Criando um DataFrame a partir de um dicionário (útil para testes locais)

dados_turma = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'Idade': [20, 22, 23, 21],
    'Nota_Final': [8.5, 9.0, 7.5, 6.5],
    'Aprovado': [True, True, True, False]
}

df = pd.DataFrame(dados_turma)
print('Exemplo de DataFrame:')
display(df) # O display renderiza uma tabela mais bonita no Jupyter do que o 'print'
```

#### 5. Seleção e Filtragem de Dados
Como selecionar colunas específicas e filtrar linhas baseadas em condições

```python
#Selecionando uma coluna (retorna uma Series)
idades = df_titanic['Age']

#Selecionando múltiplas colunas (retorna um novo df)
df_reduzido = df_titanic[['Name', 'Age', 'Survived']]

#Filtragem (querying): Selecionando apenas passageiros com mais de 60 anos
passageiros_idosos = df_titanic[df_titanic['Age'] > 60]
print('Total de passageiros com mais de 60 anos:', len(passageiros_idosos))
```

```python
#Filtragem de múltiplas condições
#Ex.: Mulheres que sobreviveram na 1° classe
filtro = (df_titanic['Sex'] == 'female') & (df_titanic['Survived'] == 1) & (df_titanic['Pclass'] == 1) #São 3 tuplas
mulheres_1a_classe = df_titanic[filtro]

print('Mulheres da 1a classe que sobreviveram', len(mulheres_1a_classe))
display(mulheres_1a_classe)
```
