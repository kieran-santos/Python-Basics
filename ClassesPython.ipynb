#### 1. Crie uma classe Aluno que armazene o nome e uma lista de notas. Implemente um método para calcular a média e outro para verificar se o aluno foi aprovado (média >= 6).

**Python docs**

* A list that is declared outside of the constructor should not be used as a class variable (atributo) because just a single list would be shared by all instances
* Declare it inside the constructor
* Instance variables are for data unique to each instance - class variables are for attributes and methods shared by all instances of the class

```python
class Aluno:
    def __init__(self, nome):
        self.nome = nome
        self.notas = [] # Cria uma lista vazia para cada aluno

    def add_nota(self, nota):
        self.notas.extend(nota)

    def media(self):
        media = sum(self.notas)/len(self.notas)
        return media
        ''' Encontrei uma questão nesse método: 
                1. Caso eu adicione as notas do aluno com append() todas de uma vez, com add_nota([8, 9, 10]), 
                as notas são adicionadas como uma lista dentro da lista de notas, sendo assim é preciso
                chamar o elemento 0 da lista self.notas - [[8, 9, 10]]
                2. Caso eu adicione utilizando append() uma por uma, com add_notas(8), add_notas(9) e etc., as notas são
                adicionadas uma por uma na lista de notas - [8, 9, 10]
                3. Posso utilizar extend() para adicionar várias de uma vez
        '''
        
    def aprovado(self):
        media = self.media()
        if media >= 6:
            return f'A média do aluno é {media} e ele foi aprovado'
        else:
            return f'A média do aluno é {media} e ele foi reprovado'

    def infos(self):
        return f'Aluno: {self.nome} | Notas: {self.notas}'


# Instanciando objetos
aluno1 = Aluno('Kieran')
aluno1.add_nota([8, 9, 10])
print(aluno1.infos())
print(aluno1.aprovado())
```

#### 2. Crie uma classe Conversor que possua um atributo privado taxa_dolar. Implemente métodos para atualizar a taxa (com validação para não aceitar valores negativos) e para converter um valor de Real para Dólar.

```python
class Conversor:
    def __init__(self, taxa_dolar):
        self.__dolar = taxa_dolar

    def atualizar(self, valor):
        if valor < 0:
            return 'O valor informado não pode ser menor que zero'
        else:
            self.__dolar = valor

    def converter(self, real):
        if real < 0:
            return 'O valor informado não pode ser menor que zero'
        else:
            valor_convertido = real / self.__dolar
            return f'Valor em reais: {real} | Valor em dólares: {valor_convertido}'


#Instanciando
valor1 = Conversor(1)
valor1.atualizar(2)
print(valor1.converter(5))
```

#### 3. Crie as classes Livro e Biblioteca. A classe Biblioteca deve conter uma lista de objetos do tipo Livro. Implemente métodos para adicionar livros e listar todos os títulos disponíveis.

```python
class Livro:
    def __init__(self, titulo):
        self.titulo = titulo


class Biblioteca:
    def __init__(self):
        self.livros = []

    def adicionar(self, livro):
        self.livros.append(livro)

    def listar(self):
        titulos = [livro.titulo for livro in self.livros] #[resultado desejado - para cada item (livro) dentro da lista de livros]
        return f'Livros: {titulos}'


#Instanciando
livro1 = Livro('Senhor dos Anéis: A Sociedade do Anel')
livro2 = Livro('O Medo à Espreita')
livro3 = Livro('Diário do Subsolo')

biblioteca = Biblioteca()
biblioteca.adicionar(livro1)
biblioteca.adicionar(livro2)
biblioteca.adicionar(livro3)


print(biblioteca.listar())
```

#### 4. Crie uma classe base Veiculo com atributos como marca e modelo. Crie subclasses Carro (com número de portas) e Moto (com cilindradas). Ambas devem herdar as características da classe pai.

```python
class Veiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

class Carro(Veiculo):
    def __init__(self, marca, modelo, portas):
        super().__init__(marca, modelo)
        self.portas = portas

    def infos(self):
        return f'{self.marca}, {self.modelo}, {self.portas}'

class Moto(Veiculo):

    def __init__(self, marca, modelo, cilindradas):
        super().__init__(marca, modelo)
        self.cilindradas = cilindradas

    def infos(self):
        return f'{self.marca}, {self.modelo}, {self.cilindradas}'
    

#Instanciando para testar
carro1 = Carro('sei la', 'não conheço', 4)
moto1 = Moto('não sei', 'não entendo', 1000)

print(carro1.infos())
print(moto1.infos())
```

#### 9. Crie uma classe Funcionario que possua um método registrar_entrada(). Utilize o módulo datetime para capturar o horário exato em que o método foi chamado e armazene isso em uma lista de logs dentro do objeto.

```python
import datetime as dt

class Funcionario:
    def __init__(self):
        self.logs = []

    def registrar_entrada(self):
        self.logs.append(dt.datetime.now())
    
    def infos(self):
        return f'Logs: {self.logs}'

#Instanciando
funcionario = Funcionario()
funcionario.registrar_entrada()
funcionario.registrar_entrada()
funcionario.registrar_entrada()
funcionario.registrar_entrada()

print(funcionario.infos())
```
