```python
import numpy as np
import matplotlib.pyplot as plt #Também depende do numpy
```

```python
produtos = ['Produto A', 'Produto B', 'Produto C']
quantidades = [50, 80, 65]

plt.bar(produtos, quantidades)
plt.title('Quantidade de Produtos em Estoque')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.show()
```

```python
np.random.seed(42)
x = np.random.rand(50) * 10 #random pra gerar numero aleatorio
y = 2 * x + np.random.randn(50) * 2

plt.scatter(x, y)
plt.title('Relação entre x e y')
plt.xlabel('Variavel X')
plt.ylabel('Variavel Y')
plt.show()
```
