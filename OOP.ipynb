## Programação Orientada a Objetos (POO) em Python

### 1. Classe e Objetos

Classe é o molde (blueprint), enquanto Objeto é a instância real desse molde

```python
class Pessoa:
    # O método __init__ é o construtor

    def __init__(self, nome, idade): #Parâmetros do método construtor
        self.nome = nome    #Self está falando com o valor que será informado pelo usuário
        self.idade = idade  #Self.nome e self.idade é propriedade da classe

    def saudar(self):
        return f'Olá, meu nome é {self.nome} e tenho {self.idade} anos.'


#Instanciando objetos
p1 = Pessoa('Kieran', 27)
print(p1.saudar())

#O self associa o valor que você dá ao objeto que você criou - faz referência ao valor de cada instãncia
#Para diferenciar o que é valor de uma instância e o que deve ser de todas as instâncias
```

### 2. Encapsulamento

Uso de atributos privados (prefixados com `__`) para proteger os dados.

```python
class ContaBancaria:
    def __init__(self, saldo_inicial):
        self.__saldo = saldo_inicial #Atributo privado

    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor

    def consultar_saldo(self):
        return f'Saldo atual: R${self.__saldo:.2f}'
    
conta = ContaBancaria(100)
conta.depositar(500)
print(conta.consultar_saldo())
```

### 3. Herança

Permite que uma classe filha herde atributos e métodos de uma classe pai.

```python
class Funcionario(Pessoa):
    def __init__(self, nome, idade, cargo):
        super().__init__(nome, idade) # Chama o construtor da classe pai
        self.cargo = cargo

    def saudar(self):
        return f"Olá, sou {self.nome}, atuo como {self.cargo}."

professor = Funcionario("Tiago", 35, "Professor")
print(professor.saudar())
```

### 4. Polimorfismo

Capacidade de diferentes classes tratarem o mesmo método de formas distintas.

```python
class Animal:
    def falar(self):
        pass

class Cachorro(Animal):
    def falar(self):
        return "Au Au!"

class Gato(Animal):
    def falar(self):
        return "Miau!"

animais = [Cachorro(), Gato()]
for animal in animais:
    print(animal.falar())
```
