## Fundamentos de Python para Machine Learning

As variáveis são tratadas como objetos e podemos chamar funções para elas

### 1. Estrutura de dados (listas, tuplas e dicionários)

```python
#Lista é mutável
frutas = ['maçã', 'banana', 'cereja']
frutas.append('uva')

#Tupla é imutável
coordenadas = (30, 20)

#Dicionário é chave: valor
aluno = {'nome': 'tiago', 'idade': 30, 'curso': 'Tecnologia'}

print(f'Lista: {frutas}')
print(f'Aluno: {aluno['nome']}')
```

### 2. Condicionais e Loops

```python
# Condicional
media = 7.5
if media >= 6:
    resultado = "Aprovado"
else:
    resultado = "Exame"

# Loop for

numeros = [1, 2, 3, 4, 5]
quadrado = [x**2 for x in numeros] # vai elevar ao quadrado cada item na lista numeros

print(f'Resultado: {quadrado}')

print("\nLista de Numeros (jeito longo)")
quadrados = []
for x in numeros:
    quadrados.append(x**2)
print(quadrados)
```

### 3. Função e Lambda

```python
def calcular_bonus(salario, percentual=10):
    return salario + (salario * percentual / 100)

#Função Lambda (anônima)
dobro = lambda x: x*2

print(f'Bônus: {calcular_bonus(3000)}')
print(f'Dobro de 5: {dobro(5)}')
```
